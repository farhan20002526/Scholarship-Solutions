```python
import requests
from bs4 import BeautifulSoup

# Fetch the web page
url = 'https://ischoolconnect.com/blog/list-of-100-graduate-schools-with-low-gpa-requirements/'
response = requests.get(url)

# Parse the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table element
table = soup.find('table')

# Initialize lists to store data
universities = []
cgpa_requirements = []

# Loop through rows in the table
for row in table.find_all('tr'):
    columns = row.find_all('td')
    if len(columns) == 2:
        university = columns[0].text.strip()
        cgpa = columns[1].text.strip()
        universities.append(university)
        cgpa_requirements.append(cgpa)

# Print the data
for i in range(len(universities)):
    print(f"University: {universities[i]}")
    print(f"CGPA Requirement: {cgpa_requirements[i]}")
    print()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Specify the URL
url = 'https://ischoolconnect.com/blog/list-of-100-graduate-schools-with-low-gpa-requirements/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table element
    tables = soup.find_all('table')

    # Initialize lists to store data
    universities = []
    cgpa_requirements = []

    # Loop through rows in the table
    for table in tables:
        for row in table.find_all('tr'):
            columns = row.find_all('td')
            if len(columns) == 2:
                university = columns[0].text.strip()
                cgpa = columns[1].text.strip()
                universities.append(university)
                cgpa_requirements.append(cgpa)

    # Create a DataFrame from the lists
    data = {
        'Universities': universities,
        'GPA': cgpa_requirements
    }
    df = pd.DataFrame(data)

    # Print the DataFrame
    print(df)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
import pandas as pd

def convert_column_to_float(df, column_name):
    """
    Convert a column in a DataFrame to float.

    Args:
        df (pd.DataFrame): The DataFrame.
        column_name (str): The name of the column to convert.

    Returns:
        pd.DataFrame: The DataFrame with the specified column converted to float.
    """
    df[column_name] = df[column_name].apply(lambda x: float(x) if pd.notna(x) else None)
    return df

# Example usage:
# Assuming you already have a DataFrame 'df' with a 'GPA' column
# Convert the 'GPA' column to float
df = convert_column_to_float(df, 'GPA')
```

```python
import pandas as pd
import re

def convert_column_to_float(df, column_name):
    """
    Convert a column in a DataFrame to float by extracting the numeric part.

    Args:
        df (pd.DataFrame): The DataFrame.
        column_name (str): The name of the column to convert.

    Returns:
        pd.DataFrame: The DataFrame with the specified column converted to float.
    """
    # Use regular expression to extract the float part
    df[column_name] = df[column_name].str.extract(r'(\d+\.\d+)').astype(float)
    return df
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the starting page
base_url = "https://unischolars.com/universities/uk?sort=popularity"

# Initialize lists to store university data
university_data = []

# Define the number of pages you want to scrape
num_pages = 5

for page in range(1, num_pages + 1):
    # Construct the URL for the current page
    url = f"{base_url}&page={page}"

    # Send an HTTP GET request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find and extract university information
        universities = soup.find_all('div', class_='uni-box')
        for university in universities:
            name = university.find('h2').text.strip()
            location = university.find('span', class_='loc').text.strip()
            tuition_fee = university.find('p', class_='fees').text.strip()
            cost_of_living = university.find('p', class_='cost').text.strip()
            country_rank = university.find('p', class_='rank').text.strip()

            # Store the data in a dictionary
            university_info = {
                'Name': name,
                'Location': location,
                'Tuition Fee': tuition_fee,
                'Cost of Living': cost_of_living,
                'Country Rank': country_rank
            }

            university_data.append(university_info)

# Print the scraped data
for university_info in university_data:
    print(university_info)
```

```python
import requests
from bs4 import BeautifulSoup

# Define the base URL
base_url = "https://unischolars.com/universities/uk?sort=popularity"

# Define headers to mimic a real browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

# Send an HTTP GET request to the URL with headers
response = requests.get(base_url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find and extract the university information
    universities = soup.find_all('div', class_='universityPageStyles_UniversityCard__2-cFL')

    for university in universities:
        name_element = university.find('h3', {'data-cy': 'university_name'})
        location_element = university.find('p', {'data-cy': 'university_location'})
        tuition_fee_element = university.find('h5', {'data-cy': 'university_tuition_fee'})
        cost_of_living_element = university.find('h5', {'data-cy': 'university_cost_of_living'})
        country_rank_element = university.find('h5', {'data-cy': 'university_country_rank'})

        # Check if elements are not None before extracting text
        if name_element:
            name = name_element.text.strip()
        else:
            name = "N/A"

        if location_element:
            location = location_element.text.strip()
        else:
            location = "N/A"

        if tuition_fee_element:
            tuition_fee = tuition_fee_element.text.strip()
        else:
            tuition_fee = "N/A"

        if cost_of_living_element:
            cost_of_living = cost_of_living_element.text.strip()
        else:
            cost_of_living = "N/A"

        if country_rank_element:
            country_rank = country_rank_element.text.strip()
        else:
            country_rank = "N/A"

        # Print the extracted information
        print(f"University Name: {name}")
        print(f"Location: {location}")
        print(f"Tuition Fee: {tuition_fee}")
        print(f"Cost of Living: {cost_of_living}")
        print(f"Country Rank: {country_rank}")
        print("\n")

else:
    print(f"Failed to retrieve data. Status code: {response.status_code} - URL: {base_url}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://unischolars.com/universities/uk?sort=popularity"

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.1000.0 Safari/537.36"
}

# Send an HTTP GET request to the URL
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all university cards on the page
    university_cards = soup.find_all('div', class_='universityPageStyles_UniversityCard__2-cFL')

    # Loop through each university card and extract information
    for card in university_cards:
        name = card.find('h3', {'data-cy': 'university_name'}).text.strip()
        location = card.find('p', {'data-cy': 'university_location'}).text.strip()
        
        # Extract cost of living, tuition fee, and country rank
        cost_of_living_elem = card.find('div', {'data-cy': 'university_cost_of_living'})
        tuition_fee_elem = card.find('div', {'data-cy': 'university_tuition_fee'})
        country_rank_elem = card.find('div', {'data-cy': 'university_country_rank'})
        
        cost_of_living = cost_of_living_elem.find('h5').text.strip() if cost_of_living_elem else 'N/A'
        tuition_fee = tuition_fee_elem.find('h5').text.strip() if tuition_fee_elem else 'N/A'
        country_rank = country_rank_elem.find('h5').text.strip() if country_rank_elem else 'N/A'

        # Print the extracted information for each university
        print(f"University Name: {name}")
        print(f"Location: {location}")
        print(f"Cost of Living: {cost_of_living}")
        print(f"Tuition Fee: {tuition_fee}")
        print(f"Country Rank: {country_rank}")
        print("\n" + "=" * 50 + "\n")

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the base URL
base_url = "https://unischolars.com/universities/uk?sort=popularity&page="

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.1000.0 Safari/537.36"
}

# Define the number of pages you want to scrape
num_pages = 24  # You can change this to the number of pages you want

# Create an empty DataFrame to store the data
data = []

# Loop through each page
for page_number in range(1, num_pages + 1):
    # Construct the URL for the current page
    url = base_url + str(page_number)

    # Send an HTTP GET request to the URL
    response = requests.get(url, headers=headers)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all university cards on the page
        university_cards = soup.find_all('div', class_='universityPageStyles_UniversityCard__2-cFL')

        # Loop through each university card and extract information
        for card in university_cards:
            name = card.find('h3', {'data-cy': 'university_name'}).text.strip()
            location = card.find('p', {'data-cy': 'university_location'}).text.strip()

            # Extract cost of living, tuition fee, and country rank
            cost_of_living_elem = card.find('div', {'data-cy': 'university_cost_of_living'})
            tuition_fee_elem = card.find('div', {'data-cy': 'university_tuition_fee'})
            country_rank_elem = card.find('div', {'data-cy': 'university_country_rank'})

            cost_of_living = cost_of_living_elem.find('h5').text.strip() if cost_of_living_elem else 'N/A'
            tuition_fee = tuition_fee_elem.find('h5').text.strip() if tuition_fee_elem else 'N/A'
            country_rank = country_rank_elem.find('h5').text.strip() if country_rank_elem else 'N/A'

            # Append the extracted data to the 'data' list
            data.append([name, location, cost_of_living, tuition_fee, country_rank])

    else:
        print(f"Failed to retrieve data from page {page_number}. Status code: {response.status_code}")

# Create a DataFrame from the collected data
df = pd.DataFrame(data, columns=["University Name", "Location", "Cost of Living", "Tuition Fee", "Country Rank"])

# Print the DataFrame
```
